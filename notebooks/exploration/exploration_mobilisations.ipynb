<span style="color:#e91e63; font-size:1em; font-weight:bold;"> 6.Nettoyage des temps d'intervention : TurnoutTimeSeconds; TravelTimeSeconds;AttendanceTimeSeconds</span>

Le nettoyage des données sur les temps d’intervention des pompiers se fera en fixant des bornes réalistes pour chaque durée :
-Le temps de départ (TurnoutTimeSeconds) doit être entre 1 seconde et 1 heure.
-Le temps de trajet (TravelTimeSeconds) doit être entre 1 seconde et 2 heures.
-Le temps de présence sur les lieux (AttendanceTimeSeconds) doit être entre 1 seconde et 4 heures.
*Les valeurs en dehors de ces intervalles (négatives, nulles, ou excessivement longues) seront considérées comme erronées. Ce nettoyage permet de conserver la majorité des données, même les cas rares ou longs, pour mieux les analyser ensuite, respectant ainsi la richesse et la diversité des données !

```python
# Proposer des bornes très larges pour inclure la quasi-totalité des données
# en ne retirant que ce qui estimpossible.
bounds_inclusive = {
    # Un temps de départ ne peut pas être nul ou négatif, et ne peut pas dépasser une certaine limite très haute (ex: 1 heure = 3600s, au-delà c'est suspect).
    "TurnoutTimeSeconds": (1, 3600),
    # Un temps de trajet ne peut pas être nul/négatif. On peut imaginer un trajet très long, mais au-delà de 2 heures (7200s) bizarre !
    "TravelTimeSeconds": (1, 7200),
    # Un temps d'assistance ne peut pas être nul/négatif.Une intervention peut durer très longtemps, mais au-delà de 3-4 heures (14400s) c'est exceptionnel maybe !.
    "AttendanceTimeSeconds": (1, 14400)
}

# Copie du dataframe pour nettoyage
df_cleaned_inclusive = df_mobilisations.copy()

# Filtrage des valeurs aberrantes très extrêmes/impossibles
for col_name, (min_val, max_val) in bounds_inclusive.items():
    df_cleaned_inclusive = df_cleaned_inclusive[(df_cleaned_inclusive[col_name] >= min_val) & (df_cleaned_inclusive[col_name] <= max_val)]

# Vérification post-nettoyage
print("Nombre de lignes après nettoyage (très inclusif) :", len(df_cleaned_inclusive))
display(df_cleaned_inclusive.describe())
```
